```python
roads_dict = {}
for index, road in roads.iterrows():
    road_name = road['road']
    points_list = []

    values = road.iloc[1:].values
    
    for i in range(0, len(values), 3):
        # Ensure we don't go out of bounds
        if i + 2 < len(values):
            name = values[i]
            lat = values[i+1]
            lon = values[i+2]
            
            # Check if the data is valid (not NaN) before adding
            if pd.notna(name) and pd.notna(lat) and pd.notna(lon):
                points_list.append({
                    'lrp': name, 
                    'lat': lat, 
                    'lon': lon
                })
    
    roads_dict[road_name] = points_list

    
```
